```python
from time import sleep, time

def inc(x):
    sleep(1)
    print(x)
    return x + 1

def second_inc(x):
    sleep(3)
    print(x)
    return x + 1

from streamz import Stream
```

```python
from dask.distributed import Client
client = Client()
```

```python
source = Stream()
x = []
source.scatter().map(inc).map(second_inc).buffer(8).gather().sink(x.append)
```
